```python
# 1. Импорт библиотек
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import requests
from io import StringIO
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

output_notebook()

# 2. Получение данных с Википедии (Гидроэнергетика)
url = "https://ru.wikipedia.org/wiki/%D0%93%D0%B8%D0%B4%D1%80%D0%BE%D1%8D%D0%BD%D0%B5%D1%80%D0%B3%D0%B5%D1%82%D0%B8%D0%BA%D0%B0"
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36"}

response = requests.get(url, headers=headers)
tables = pd.read_html(StringIO(response.text))

print(f"Найдено таблиц: {len(tables)}")

# 3. Ищем таблицу с данными о ГЭС (обычно "Крупнейшие ГЭС" или "Страны по гидроэнергии")
df_hydro = None
name_col_idx = -1
power_col_idx = -1
is_power_in_gw = False

for i, table in enumerate(tables):
    if len(table.columns) >= 2:
        # Check column headers (first row of the table)
        headers = [str(col).lower() for col in table.columns]

        # Look for name-related columns
        potential_name_cols = [idx for idx, header in enumerate(headers) if any(kw in header for kw in ['название', 'гэс', 'станция', 'страна', 'территория'])]
        # Look for power-related columns
        potential_power_cols = [idx for idx, header in enumerate(headers) if any(kw in header for kw in ['мощность', 'мвт', 'гвт', 'квт'])]

        if potential_name_cols and potential_power_cols:
            df_hydro = table
            name_col_idx = potential_name_cols[0] # Take the first found
            power_col_idx = potential_power_cols[0] # Take the first found

            # Check if power is in GW
            if 'гвт' in headers[power_col_idx]:
                is_power_in_gw = True

            print(f"\nИспользуем таблицу {i} (по заголовкам): {table.shape}")
            break

if df_hydro is None:
    # Fallback to tables[1] as it seems to be 'Мощность, ГВт' from kernel state
    # This is a heuristic and might need manual adjustment if Wikipedia page structure changes
    if len(tables) > 1: # Ensure tables[1] exists
        df_hydro = tables[1]
        name_col_idx = 0 # Assuming 'Территория' or similar is in the first column
        power_col_idx = 1 # Assuming 'Мощность, ГВт' is in the second column
        if 'гвт' in str(df_hydro.columns[power_col_idx]).lower():
            is_power_in_gw = True
        print(f"\nИспользуем таблицу 1 (fallback, Мощность в ГВт): {df_hydro.shape}")
    else:
        print("Не удалось найти подходящую таблицу с данными о ГЭС.")
        # Further code will likely error, which is acceptable if data can't be found.

if df_hydro is not None:
    # Clean up name column: remove potential numerical indices, etc.
    df_clean = pd.DataFrame()
    df_clean['Name'] = df_hydro.iloc[:, name_col_idx].astype(str).str.replace(r'\[\d+\]', '', regex=True).str.strip()
    df_clean['Power_Raw'] = df_hydro.iloc[:, power_col_idx].astype(str)

    # Очищаем числа от пробелов, запятых и нецифровых символов
    df_clean['Power_MW'] = df_clean['Power_Raw'].str.replace(r'[^\d,\.]', '', regex=True) # Allow decimal points
    df_clean['Power_MW'] = df_clean['Power_MW'].str.replace(',', '.') # Replace comma with dot for decimals
    df_clean['Power_MW'] = pd.to_numeric(df_clean['Power_MW'], errors='coerce')

    # If the original unit was GW, convert to MW
    if is_power_in_gw:
        df_clean['Power_MW'] = df_clean['Power_MW'] * 1000

    # Удаляем строки без мощности и нулевую мощность
    df_clean = df_clean.dropna(subset=['Power_MW'])
    df_clean = df_clean[df_clean['Power_MW'] > 0]

    # Remove rows where 'Name' might be a header from an inconsistent table or empty
    df_clean = df_clean[df_clean['Name'].str.len() > 1]
    df_clean = df_clean[~df_clean['Name'].str.lower().str.contains('мощность')] # Avoid header-like names

    # Берём топ-10 для наглядности (уже отсортированный по убыванию мощности)
    df_top10 = df_clean.sort_values('Power_MW', ascending=False).head(10).copy()

    # Rename columns to Russian
    df_top10.rename(columns={
        'Name': 'Название',
        'Power_MW': 'Мощность (МВт)',
        'Power_GW': 'Мощность (ГВт)',
        'Category': 'Категория'
    }, inplace=True)

    print("\nТоп-10 ГЭС по установленной мощности:")
    print(df_top10[['Название', 'Мощность (МВт)']])

    # ====== 5. ВЫПОЛНЕНИЕ ПУНКТОВ ЗАДАНИЯ 4.3 ======

    # 4.3.1 Сортировка
    print("\n1. Сортировка по мощности (убывание):")
    print(df_top10.sort_values('Мощность (МВт)', ascending=False)[['Название', 'Мощность (МВт)']])

    # 4.3.2 Добавление столбца с новыми данными
    df_top10['Мощность (ГВт)'] = df_top10['Мощность (МВт)'] / 1000
    df_top10['Категория'] = np.where(
        df_top10['Мощность (МВт)'] > 15000, 'Гигант (>15 ГВт)',
        np.where(df_top10['Мощность (МВт)'] > 5000, 'Крупная (5-15 ГВт)',
        np.where(df_top10['Мощность (МВт)'] > 1000, 'Средняя (1-5 ГВт)', 'Малая (<1 ГВт)'))
    )
    print("\n2. Добавлены столбцы Мощность (ГВт) и Категория:")
    print(df_top10[['Название', 'Мощность (МВт)', 'Мощность (ГВт)', 'Категория']])

    # 4.3.3 Выборка по условию
    print("\n3. ГЭС мощностью > 10 000 МВт:")
    print(df_top10[df_top10['Мощность (МВт)'] > 10000][['Название', 'Мощность (МВт)']])

    # 4.3.4 Выборка по номерам (iloc)
    print("\n4. Первые 3 строки, столбцы Название и Мощность (МВт):")
    print(df_top10.iloc[0:3, df_top10.columns.get_loc('Название') : df_top10.columns.get_loc('Мощность (МВт)')+1]) # Use get_loc for robustness

    # 4.3.5 Группировка и статистика
    print("\n5. Статистика по категориям (сумма, среднее, количество):")
    stats = df_top10.groupby('Категория')['Мощность (МВт)'].agg(['sum', 'mean', 'count'])
    print(stats)

    # 4.3.6 Гистограмма
    plt.figure(figsize=(14, 6))
    bars = plt.bar(df_top10['Название'], df_top10['Мощность (МВт)'], color='teal', edgecolor='black')
    plt.title('Мощность крупнейших ГЭС мира (топ-10)', fontsize=14)
    plt.xlabel('Название ГЭС')
    plt.ylabel('Мощность (МВт)')
    plt.xticks(rotation=45, ha='right')
    for bar in bars:
        plt.text(bar.get_x() + bar.get_width()/2., bar.get_height() + 300,
                 f'{bar.get_height():.0f}', ha='center', va='bottom', fontsize=8)
    plt.grid(axis='y', alpha=0.3)
    plt.tight_layout()
    plt.show()

    # 4.3.7 Круговая диаграмма
    plt.figure(figsize=(10, 10))
    plt.pie(df_top10['Мощность (МВт)'], labels=df_top10['Название'], autopct='%1.1f%%',
            startangle=90, colors=plt.cm.Set3.colors)
    plt.title('Доля крупнейших ГЭС в общей мощности (топ-10)', fontsize=14)
    plt.axis('equal')
    plt.tight_layout()
    plt.show()

    # 4.3.8 Интерактивный график (Bokeh)
    df_interactive = df_top10.sort_values('Мощность (МВт)', ascending=True)

    p = figure(y_range=df_interactive['Название'].tolist(), height=500, width=900,
               title="Интерактивный график: Мощность ГЭС (топ-10)",
               x_axis_label="Мощность (МВт)", y_axis_label="Название ГЭС",
               tools="pan,wheel_zoom,reset,hover,save",
               tooltips=[("Название ГЭС", "@y"), ("Мощность (МВт)", "@right")])

    p.hbar(y=df_interactive['Название'], right=df_interactive['Мощность (МВт)'],
           height=0.7, color="steelblue", alpha=0.8)
    p.grid.grid_line_color = "gray"
    p.grid.grid_line_alpha = 0.3
    show(p)
else:
    print("Не удалось построить графики, так как данные не были найдены или обработаны корректно.")
```
